### Junção, Mescla e Concatenação 🥇

#### Inner Join

<pre>Faça a mescla do DataFrame <b>df_rais</b>, com informações das trabalhadoras, com o DataFrame <b>df_pib_pc</b>, com informações de PIB per capita das Unidades da Federação, de forma que cada observação referente a uma trabalhadora seja acrescentada a informação de PIB per capita da UF em que ela trabalha.</pre>

👉 dica: para mesclar dois DataFrames, use o método `merge` do Pandas 🐼, usando ambos os DataFrames como parâmetros, bem como utilizando os valores apropriados para os parâmetros `how`, `left_on` e `right_on`.

<details>
  <summary>Resposta</summary>

<br/>

```python
pd.merge(df_rais, df_pib_pc, how = "inner", left_on="uf", right_on="UF")
```

#### Inner Join, Left Outer Join, Right Outer Join, Outer Join

<pre>Similarmente, faça a combinação do DataFrame <b>df_amostra</b> com o dataframe <b>df_pib_pc</b>, de forma que cada observação referente a uma trabalhadora seja acrescentada a informação de PIB per capita da UF em que ela trabalha.</pre>

<pre>Qual tipo de mescla você deve usar e qual tipo de mescla você não deve usar?</pre>

<details>
  <summary>Resposta</summary>

<br/>

Observe que temos informações sobre apenas 5 trabalhadoras, enquanto temos informações de PIB per capita de 27 Unidades da Federação. Portanto, ao menos 22 Unidades da Federação não terão trabalhadoras dessas UF no DataFrame `df_amostra`. 

Se utilizarmos o `left join` ou o `inner join` nos DataFrames `df_amostra` e `df_pib_pc`, nessa ordem, nossa mescla combinará os dados de todas as trabalhadoras com suas respectivas UF de trabalho, incluindo (no resultado final) apenas os dados referentes a essas UF. Portanto, ao menos 22 UFs não terão seu PIB per capita presente no resultado final.

No entanto, se utilizarmos o `right join` ou o `outer join` nos DataFrames `df_amostra` e `df_pib_pc`, nessa ordem, nossa mescla combinará os dados de PIB per capita de todas as UF com os dados de trabalhadoras de no máximo 5 UF, apresentando linhas com valores ausentes em todas as colunas, exceto `PIB_PC`, para as UF que não têm qualquer trabalhadora.

Portanto, para atender ao objetivo acima, você pode utilizar o `left join` ou o `inner join` nos DataFrames `df_amostra` e `df_pib_pc`, nessa ordem, bem como não pode utilizar o `right join` ou o `outer join` nos DataFrames `df_amostra` e `df_pib_pc`, nessa ordem.
