# Tratamento dos Dados 🪢

<pre>A base de indicadores criminais por município e ano do Estado do Rio Grande do Sul é publicada no formato Excel. Por meio de raspagem de dados, obtivemos os dados referentes aos anos de 2012 até a presente data.</pre>

<pre>Execute as duas células abaixo para condensar todos os anos da base de indicadores criminais por município do Estado do Rio Grande do Sul em um único DataFrame, de nome <b>df_crimes</b>.</pre>

````python
!pip install --upgrade openpyxl --quiet
!pip install --upgrade pandas --quiet
````

````python
import pandas as pd

dfs = []

xlsx = pd.ExcelFile('https://raw.githubusercontent.com/heliomacedofilho/idp_fundamentos_2026/refs/heads/main/day%2003/Parte%2001/exercises/Nivel%203/data/crimes_rs_municipios.xlsx')

anos = [year for year in xlsx.sheet_names if all(c.isdigit() for c in year)]

for ano in anos:
    df = pd.read_excel(xlsx, sheet_name=ano)
    df['Ano'] = ano
    dfs.append(df)

df_crimes = pd.concat(dfs)
df_crimes.head()
````

## Junção, Mescla e Concatenação 🥇

### Junção

<pre>Vamos calcular a taxa de homicídios por habitante para cada um dos municípios do Rio Grande do Sul.</pre>

<pre>Para isso, será necessário juntar os DataFrames <b>df_homicidios</b> e <b>df_populacao</b>, utilizando um índice em comum como chave da junção.</pre>

<pre>No entanto, note que apenas o DataFrame <b>df_homicidios</b> utiliza o nome de Município como índice. Isto não ocorre para o DataFrame <b>df_populacao</b>.</pre>

<pre>Portanto, para que seja possível realizar a junção entre os DataFrames, defina a coluna <b>Município</b> como índice do DataFrame <b>df_populacao</b>.</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python
df_populacao = df_populacao.set_index('Município')
```

<br/>

</details>

<pre>Agora, faça a junção dos DataFrames <b>df_homicidios</b> e <b>df_populacao</b>, utilizando o índice (em comum) como chave da junção, e atribua o resultado para o DataFrame <b>df_tx_de_homicidio</b>.</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python
df_tx_de_homicidio = df_homicidios.join(df_populacao)
```

<br/>

</details>

<pre>Veja que o DataFrame <b>df_tx_de_homicidio</b> apresenta o número de homicídios dolosos por ano, bem como a totalização de cada município.</pre>

<pre>Como precisamos apenas da totalização de cada município, mantenha apenas as colunas <b>Total/Município</b> e <b>Pop. Estimada</b> no DataFrame <b>df_tx_de_homicidio</b>.</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python
df_tx_de_homicidio = df_tx_de_homicidio[['Total/Município', 'Pop. Estimada']]
```

<br/>

</details>

<pre>Agora, crie a coluna <b>Taxa de Homicídio Doloso</b> no DataFrame <b>df_tx_de_homicidio</b>, obtida por operação de divisão entre <b>Total/Município</b> e <b>Pop. Estimada</b>.</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python
df_tx_de_homicidio['Taxa de Homicídio Doloso'] = df_tx_de_homicidio['Total/Município'] / df_tx_de_homicidio['Pop. Estimada']
```

<br/>

</details>

<pre>Como a totalização de cada município e a população estimada não se fazem mais necessárias, mantenha apenas a coluna <b>Taxa de Homicídio Doloso</b> no DataFrame <b>df_tx_de_homicidio</b>.</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python
df_tx_de_homicidio = df_tx_de_homicidio[['Taxa de Homicídio Doloso']]
```

<br/>

</details>

<pre>Por fim, apresente todos os valores do DataFrame <b>df_tx_de_homicidio</b> com valores percentuais e precisão de duas casas decimais.</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python
df_tx_de_homicidio.map(lambda x: f'{x:.2%}')
```
