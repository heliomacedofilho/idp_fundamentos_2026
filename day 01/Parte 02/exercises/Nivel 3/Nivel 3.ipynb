````python
import csv

tabela = []

with open('data/Base-COFOG-2022.csv') as csvfile:
    for linha in csv.reader(csvfile):
        tabela.append(linha)
````

<pre>A célula abaixo realiza o cálculo acima para você. Veja que <b>NumPy</b> sempre será muito mais rápido.</pre>

````python
import time

def tempo_de_calculo(despesas):
    start = time.time()
    desp_total = sum(despesas)
    end = time.time()
    return end-start

tempo_de_calculo_com_lista = tempo_de_calculo(despesas)
tempo_de_calculo_com_array = tempo_de_calculo(despesas_numpy)

f'NumPy foi {round(tempo_de_calculo_com_array/tempo_de_calculo_com_lista)-1} vezes mais rápido do que python nativo.'
````
