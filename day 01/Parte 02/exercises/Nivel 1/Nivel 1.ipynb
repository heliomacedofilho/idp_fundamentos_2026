<pre>Por fim, produza a média salarial de homens e mulheres com um comando no <b>DataFrame</b> <b>df</b>.</pre>

👉 dica: use o <a href='https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html'>groupby</a>.

<details>
  <summary>Resposta</summary>


`Racional`: para agrupar as linhas de acordo com as possibilidades de valores (`MASCULINO` e `FEMININO`) da coluna  `genero`, utilize o método `groupby()` do `DataFrame` `df`, passando `genero` como parâmetro. Na sequência, selecione a coluna que se quer aplicar a função de agregação (`salario`) e aplique a função de agregação desejada (`mean()`).
