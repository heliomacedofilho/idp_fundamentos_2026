#### Preenchendo Dados Ausentes

<pre>Agora, vamos falar das possíveis estratégias para tratar dados ausentes.</pre>

<pre>A melhor solução, nem sempre viável, é solicitar ao responsável pela disponibilização da base de dados que apresente os dados completos.</pre>

<pre>Caso isso não seja possível ou caso os valores estejam ausentes na base original, pode ser inevitável que haja alguma imprecisão na análise.</pre>

<pre>Dependendo do contexto, pode ser aceitável imputar valores aos dados ausentes e isso pode ser feito por diversos métodos.</pre>

<pre>Vamos ilustrá-los nos dados referentes à <b>AGENCIA NACIONAL DE SAUDE SUPLEMENTAR</b>. Portanto, execute a célula abaixo.</pre>

````python
df_inativos_ans = df_inativos[df_inativos["OrgN2"] == "AGENCIA NACIONAL DE SAUDE SUPLEMENTAR"].copy()
df_inativos_ans
````

<pre>    1 - Utilize o método <b>fillna</b> (do DataFrame <b>df_inativos_ans</b>) para substituir os valores ausentes (<b>None</b>) por <b>0</b>;</pre>

<details>
  <summary>Resposta</summary>

<br/>

```python

```

<br/>

</details>

<pre>    2 - Alternativamente, utilize o valor da linha anterior para preencher os valores ausentes, com o método <b>ffill</b> do DataFrame <b>df_inativos_ans</b>;</pre>

🚨 Similarmente, você pode utilizar o valor da linha posterior para preencher os valores ausentes, utilizando o método <b>bfill</b>. 🚨

  <summary>Resposta</summary>

<br/>


</details>

<pre>    3 - Por fim, utilize valores existentes para calcular um valor intermediário, com o método <b>interpolate</b> do DataFrame <b>df_inativos_ans</b>.</pre>

👉 dica: consulte a documentação do <a href='https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.interpolate.html'>interpolate</a>.

## Reorganização 🗂️ dos Dados 🎲

<pre>A biblioteca Pandas 🐼 permite a leitura de diversos formatos de arquivos, inclusive planilhas Excel, muito utilizadas para o armazenamento e publicação de pequenas bases de dados.</pre>

<pre>A base de <a href="https://ssp.rs.gov.br/indicadores-criminais">indicadores criminais por município e ano do Estado do Rio Grande do Sul</a> é publicada no formato Excel. Execute as duas células abaixo para obter os Indicadores criminais geral e por município do ano de 2024.</pre>

````python
!pip install --upgrade openpyxl --quiet
````

````python
from bs4 import BeautifulSoup
from requests import get
from urllib.parse import urljoin

base = 'https://www.ssp.rs.gov.br/indicadores-criminais'
html = get(base).content
soup = BeautifulSoup(html, 'html.parser')

# identifica a URL da planilha de 2024 que tem dados de crimes no RS
urls = [urljoin(base, element.get('href')) for element in soup.find_all('a') if element.get('href') and all(substring in element.get('href') for substring in ['geral-e', 'municipio', '-2024'])]

#salva o arquivo no sistema de arquivos
xlsx = get(urls[0])
if xlsx.status_code == 200:
    with open('data/indicadores_criminais_rs_2024.xlsx', mode='wb') as xlsx_file:
        xlsx_file.write(xlsx.content)
````

<pre>A planilha armazenada em <b>data/indicadores_criminais_rs_2024.xlsx</b> contém dados consolidados na planilha denominada <b>Geral</b> e dados para cada município com periodicidade anual e mensal.</pre>
